```python
#코로나 데이터
import pandas as pd
import numpy as np
df_covid19= pd.read_csv('data/owid-covid-data.csv')

print('row count:', len(df_covid19))
df_covid19
```

Variable	:  Description

*   total_cases	: Total confirmed cases of COVID-19 
* total_vaccinations : Total number of COVID-19 vaccination doses administered (투여된 수)
*   people_fully_vaccinated: Total number of people who received all doses prescribed by the vaccination protocol (예방 접종 프로토콜에 명시된 모든 용량을 접종받은 총 사람 수)



```python
df_tcvx= df_covid19[['continent','location', 'date', 'total_cases', 'new_vaccinations', 'people_fully_vaccinated', 'new_cases']]
df_tcvx = df_tcvx.fillna(0)
df_tcvx.tail()
```

```python
### 코로나 백신 접종자수 불러오기
c_data = df_tcvx.groupby(['date']).sum()
corona_data = c_data[['new_cases','new_vaccinations']]
corona_data.index = pd.to_datetime(corona_data.index)
corona_data
```

코로나 데이터의 경우 NA값은 관측되지 않은 것이라고 판단, 자체적으로 0값으로 치환하였다.
